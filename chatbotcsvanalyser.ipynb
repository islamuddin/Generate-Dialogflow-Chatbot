```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('qna_chitchat_the_friend.tsv',delimiter='\t',encoding='utf-8')
df.head(50)
```

```python
df=df[['Question','Answer','Source']]
df.head(10)
```

```python
df.columns=['User Says','Text Response','Intent']
df.head(10)
```

```python
for index,row in df.iterrows():
    row['Intent']='friendlychat'+str(index)
```

```python
df.head(10)
```

```python
df.to_csv('friendlybot.csv')
```

```python
indices = []
for i,row in df.iterrows():
    indices.append(i)
```

```python
i = 0
j = 0
num_data_points = df.shape[0]
while i < num_data_points and j < num_data_points:
    previous_i = i
    a = str(df['Text Response'].loc[indices[i]]).split()
    # search for the similar products sequentially 
    j = i+1
    while j < num_data_points:
        b = str(df['Text Response'].loc[indices[j]]).split()
        if a==b:
            df['Intent'].loc[indices[j]]=df['Intent'].loc[indices[i]]
            df['Text Response'].loc[indices[j]]=np.NaN
        j=j+1;
    j=i+1;
    i=i+1
```
